# Checking for outliers 

```python
plt.figure(figsize=(27,20))
plotnum = 1
for i in df.columns:
    if plotnum <= 17 :
        ax = plt.subplot(6,3,plotnum)
        sns.boxplot(df[i])
        plt.xlabel(i , fontsize=14)
    plotnum +=1
plt.show()    
```

we see that columns Run Scored , Earned Runs Average , Shut Out , saves and errors have out liers

```python
# using z score method
from scipy.stats import zscore
```

```python
outliers = df[['Run Scored' ,'Earned Runs Average' ,'Shut Out', 'saves', 'errors']]
z = np.abs(zscore(outliers))
```

```python
df1 = df[(z<3).all(axis=1)]
df1.shape
```

we removed one row

```python
# use IQR method
out = df[['Run Scored' ,'Earned Runs Average' ,'Shut Out', 'saves', 'errors']]

#1st percentile
q1 = out.quantile(0.25)

#3rd percentile
q3 = out.quantile(0.75)

IQR = q3 - q1
df3 = df[~((df < (q1 - 1.5*IQR)) | (df > (q3 + 1.5*IQR))).any( axis =1)]
```

```python
df3.shape
```

we removed 10 outliers s IQR is the best method

```python
plt.figure(figsize=(27,20))
plotnum = 1
for i in df3.columns:
    if plotnum <= 17 :
        ax = plt.subplot(6,3,plotnum)
        sns.boxplot(df3[i])
        plt.xlabel(i , fontsize=14)
    plotnum +=1
plt.show()    
```

We remved utliers as possible as we could

# Scaling Data

```python
x = df3.drop('wins' , axis=1)
y= df3['wins']
print('shape of features is ',x.shape)
print('shape of target variable is ',y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled = pd.DataFrame(scaler.fit_transform(x) , columns = x.columns )
x_scaled
```

# check multi collinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(x_scaled.values , i) for i in range(len(x_scaled.columns))]
vif['column'] = x_scaled.columns
vif
```

There is a multicllinearity problem in many columns so we will remove it

# Principle Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit_transform(x_scaled)
```

```python
#now let us see the screen to check the best components
plt.figure(figsize=(15,10))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('Variance %')
plt.title('Explained Variance Ratio')
plt.grid(axis='both')
plt.xticks(np.arange(0,16,1))
plt.show()
```

As we see the number of 8 components get variance of 95% so will choose it

```python
pca = PCA(n_components=8)
new_component = pca.fit_transform(x_scaled)
components = pd.DataFrame(new_component , columns = np.arange(8))
components
```

```python
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(components , i) for i in range(len(components.columns))]
vif['column'] = components.columns
vif
```

We slved the problem of multicollinearity

# Building Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
# find the best randm stats
accu = []
stat = []
for i in range(1,200):
    x_train , x_test , y_train , y_test = train_test_split(components , y , test_size=0.3 , random_state=i)
    lr = LinearRegression()
    lr.fit(x_train,y_train)
    y_pred = lr.predict(x_test)
    accuracy = r2_score(y_test , y_pred)
    accu.append(accuracy)
    stat.append(i)
```

```python
rand = pd.DataFrame()
rand['accuracy'] = accu
rand['stats'] = stat
rand
```

```python
rand[rand['accuracy'] == rand['accuracy'].max()]
```

The best accuracy is 94.284 % when number of states is 35

```python
x_train , x_test , y_train , y_test = train_test_split(components , y , test_size=0.3 , random_state=35)
```

```python
from sklearn.ensemble import RandomForestRegressor , GradientBoostingRegressor , ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Lasso , Ridge
from sklearn.svm import SVR
```

```python
rfr = RandomForestRegressor()
gbr = GradientBoostingRegressor()
et = ExtraTreesRegressor()
dtr = DecisionTreeRegressor()
knn = KNeighborsRegressor()
las = Lasso()
ri = Ridge()
svr = SVR()
```

```python
models = {'LinearRegression':lr ,'RandomForestRegressor':rfr ,'GradientBoostingRegressor':gbr ,
        'ExtraTreesRegressor':et , 'DecisionTreeRegressor':dtr ,
          'KNeighborsRegressor':knn , 'Lasso':las , 'Ridge':ri ,'SVR':svr}
for name , m in models.items():
    m.fit(x_train , y_train)
    pred_train = m.predict(x_train)
    pred_test = m.predict(x_test)
    print('model name is ',name)
    print('accuracy score on test data is ','{:.2f}'.format(r2_score(y_test,pred_test)*100),' %')
    print('accuracy score on train data is ','{:.2f}'.format(r2_score(y_train,pred_train)*100),' %')
    print('mean absolute error is ',mean_absolute_error(y_test,pred_test))
    print('mean squared error is ',mean_squared_error(y_test,pred_test))
    print('root of mean squared error is ',np.sqrt(mean_squared_error(y_test,pred_test)))
    print('\n')
```

Ridge model has the best accuracy
