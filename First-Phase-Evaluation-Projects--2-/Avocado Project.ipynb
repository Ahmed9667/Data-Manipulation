# Checking for outliers

```python
plt.figure(figsize=(25,20))
plotnum = 1
for i in df.columns:
    if plotnum <= 14 :
        ax = plt.subplot(5,3,plotnum)
        if df[i].dtypes == 'object' :
            continue
        sns.boxplot(df[i])
        plt.xlabel(i , fontsize=12)
    plotnum += 1
plt.show()    
```

- all columns contains continuous numerical data have outliers so we will remove them
- columns year and Month have categorical data and have no outliers

```python
# use Zscre method
from scipy.stats import zscore
```

```python
outliers = df[['AveragePrice', 'Total Volume', '4046', '4225', '4770', 'Total Bags',
       'Small Bags', 'Large Bags', 'XLarge Bags']]
z = np.abs(zscore(outliers))
```

```python
df1 = df[(z<3).all(axis=1)]
df1.shape
```

```python
print('percentage of loss is ',((df.shape[0] - df1.shape[0])/df.shape[0])*100,' %')
```

```python
#use IQR method
out= df[['AveragePrice', 'Total Volume', '4046', '4225', '4770', 'Total Bags',
       'Small Bags', 'Large Bags', 'XLarge Bags']]

q1 = out.quantile(0.25)  #first percentile
q3 = out.quantile(0.75)  #third percentile
IQR = q3-q1

df2 = df[~((df < (q1 - 1.5*IQR) ) | (df > (q3 + 1.5*IQR))).any(axis=1)]
df2.shape
```

```python
print('percentage of loss is ',((df.shape[0] - df2.shape[0])/df.shape[0])*100,' %')
```

IQR method is the best method to remove outliers

```python
# now let us check outliers
plt.figure(figsize=(25,20))
plotnum = 1
for i in df2.columns:
    if plotnum <= 14 :
        ax = plt.subplot(5,3,plotnum)
        if df[i].dtypes == 'object' :
            continue
        sns.boxplot(df2[i])
        plt.xlabel(i , fontsize=12)
    plotnum += 1
plt.show()    
```

we reduced outliers as possible

we reduced skewness and all data looks normally distributed so the data is ready to be performed

# Building Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
accuracy = []
stat = []
lr = LinearRegression()
for i in range(1,200):
    x_train , x_test , y_train , y_test = train_test_split(x_scaled,y,random_state = i , test_size=0.3)
    lr.fit(x_train , y_train)
    y_pred = lr.predict(x_test)
    accu = r2_score(y_test , y_pred)
    accuracy.append(accu)
    stat.append(i)
```

```python
random = pd.DataFrame()
random['accuracy'] = accuracy
random['no of random states'] = stat
random
```

```python
random[random['accuracy'] == random['accuracy'].max()]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_scaled , y , random_state = 102 , test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor , GradientBoostingRegressor , ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Lasso , Ridge
from sklearn.svm import SVR
```

```python
rfr = RandomForestRegressor()
gbr = GradientBoostingRegressor()
et = ExtraTreesRegressor()
dtr = DecisionTreeRegressor()
knn = KNeighborsRegressor()
las = Lasso()
ri = Ridge()
svr = SVR()
```

```python
models = {'LinearRegression':lr ,'RandomForestRegressor':rfr ,'GradientBoostingRegressor':gbr ,
        'ExtraTreesRegressor':et , 'DecisionTreeRegressor':dtr ,
          'KNeighborsRegressor':knn , 'Lasso':las , 'Ridge':ri ,'SVR':svr  }
for name , m in models.items():
    m.fit(x_train , y_train)
    pred_train= m.predict(x_train)
    pred_text = m.predict(x_test)
    print('model name is : ',name)
    print('accuracy score on test data is: ','{:.2f}'.format(r2_score(y_test ,pred_text)*100),' %')
    print('accuracy score on train data is: ','{:.2f}'.format(r2_score(y_train ,pred_train)*100),' %')
    print('mean absolute error is : ',mean_absolute_error(y_test ,pred_text))
    print('mean squared error is : ',mean_squared_error(y_test ,pred_text))
    print('Root mean squared is ',np.sqrt(mean_squared_error(y_test ,pred_text)))
    print('\n')
```

ExtraTreesRegressor mdel has the highest accuracy

```python
from sklearn.model_selection import cross_val_score
```

```python
models = {'RandomForestRegressor':rfr ,'GradientBoostingRegressor':gbr ,
        'ExtraTreesRegressor':et , 'DecisionTreeRegressor':dtr ,
          'KNeighborsRegressor':knn , 'Lasso':las , 'Ridge':ri ,'SVR':svr }
for name , m in models.items():
    score = cross_val_score(m , x_scaled , y , scoring='r2' , cv=5)
    mean = score.mean()
    print('model name : ',name)
    print('score is : ',score)
    print('mean is : ',mean)
    print('Difference between r2 score and cross validation score is : ',r2_score(y_test,m.predict(x_test)) - mean)
```
